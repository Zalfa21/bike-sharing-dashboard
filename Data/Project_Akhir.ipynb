## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah terdapat tren musiman dalam jumlah peminjaman sepeda?

```python
df['month'] = pd.to_datetime(df['dteday']).dt.month
df['month_name'] = df['month'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%B'))

plt.figure(figsize=(12,6))
sns.barplot(x='month_name', y='cnt', data=df, estimator='mean',
            order=pd.date_range('2023-01-01', periods=12, freq='M').strftime('%B'))
plt.title('Rata-rata Jumlah Peminjaman Sepeda per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peminjaman')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh kondisi cuaca terhadap jumlah peminjaman sepeda?

```python
weather_labels = {
    1: 'Cerah/Berawan',
    2: 'Berkabut',
    3: 'Hujan Ringan',
    4: 'Cuaca Buruk'
}
df['weathersit_label'] = df['weathersit'].map(weather_labels)

plt.figure(figsize=(10,6))
sns.boxplot(x='weathersit_label', y='cnt', data=df)
plt.title('Distribusi Peminjaman Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Peminjaman')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight:**
- Jumlah peminjaman sepeda meningkat selama musim panas (JuniAgustus).
- Cuaca cerah atau berawan cenderung menghasilkan lebih banyak peminjaman sepeda.

## Conclusion

- Pada musim panas, peminjaman sepeda mengalami peningkatan signifikan.
- Jumlah peminjaman menurun drastis saat cuaca buruk.
